```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
import random
from functools import reduce
```

```python
def get_chapters_from_title(tit):
    link = df_title_laws['Title_Link'][df_title_laws['Title']  == tit].values[0]

    url_title_laws = url + link
    r_title_laws = requests.get(url_title_laws)
    soup_title_laws = bs(r_title_laws.content)

    title_codes_listing =soup_title_laws.find(class_ = 'codes-listing')
    title_laws = {}
    for a_tag in title_codes_listing.find_all('a'):
        text = a_tag.get_text()
        chapter = text.split('-')[1].strip()
        descript = text.split('-')[2].strip()
        link = a_tag.get('href')
        title_laws[chapter] = [descript,link]


    df_chapters = pd.DataFrame(title_laws)
    df_chapter_laws = df_chapters.T.reset_index()
    df_chapter_laws.columns = ['Chapter','Chapter_Description', 'Chapter_Link']
    df_chapter_laws.insert(loc = 0,
              column = 'Title',
              value = tit)
    return df_chapter_laws


def get_section_from_title_chapter(t,chap):
    df_chapters = get_chapters_from_title(t)
    chapter_link = df_chapters['Chapter_Link'][df_chapters['Chapter']== chap].values[0]
    url_chapter_laws = url + chapter_link
    r_chapter_laws = requests.get(url_chapter_laws)
    soup_chapter_laws = bs(r_chapter_laws.content)

    chapter_codes_listing = soup_chapter_laws.find(class_ = 'codes-listing')
    chapter_laws = {}
    for a_tag in chapter_codes_listing.find_all('a'):
        text = a_tag.get_text()
        section = text.split('-')[2].strip()
        descript = text.split('-')[3].strip()
        link = a_tag.get('href')
        chapter_laws[section] = [descript,link]

    df_section = pd.DataFrame(chapter_laws)
    df_section_laws = df_section.T.reset_index()
    df_section_laws.columns = ['Section','Section_Description', 'Section_Link']
    df_section_laws.insert(loc = 0,
              column = 'Chapter',
              value = chap)
    df_section_laws.insert(loc = 0,
              column = 'Title',
              value = t)
    return df_section_laws
#get_chapters_from_title('11')
#get_section_from_title_chapter('11','66')
#
```

```python
url = 'https://law.justia.com'
r = requests.get(url)
soup = bs(r.content)

state_box = soup.find(class_ = 'panel-pane pane-block pane-justia-laws-listings-2')
states_links = {}
for a_tag in state_box.find_all('a'):
        text = a_tag.get_text().lower()
        link = a_tag.get('href')
        states_links[text] = link
        
focus = 'Rhode Island'
year = '2021'
state_url = url + states_links.get(focus.lower() + ' law')
r_state = requests.get(state_url)
soup_state = bs(r_state.content)

# This part is tricky becuase we aren't sure what table to use
soup_state.find('ul')
primary_content = soup_state.find(class_ = 'primary-content')

# hopefully what we want is in the first undordered list
# then we can take the links like from above
laws_of_interest = primary_content.find('ul').find_all('a') 
state_laws_year = {}
for a_tag in laws_of_interest:
    law_year = a_tag.get_text().split(' ')[0]
    link = a_tag.get('href')
    state_laws_year[law_year] = link
    
year_laws_url = url + state_laws_year.get(year)
r_year_laws = requests.get(year_laws_url)
soup_year_laws = bs(r_year_laws.content)

codes_listing = soup_year_laws.find(class_ = 'codes-listing')
general_laws = {}
for a_tag in codes_listing.find_all('a'):
    text = a_tag.get_text().split("-")
    title = text[0].split(' ')[1]
    descript = text[1]
    link = a_tag.get('href')
    general_laws[title] = [descript, link]
    
df_titles = pd.DataFrame(general_laws)

df_title_laws = df_titles.T.reset_index()
df_title_laws.columns = ['Title','Title_Description','Title_Link']

title = '11'
link = df_title_laws['Title_Link'][df_title_laws['Title']  == title].values[0]

url_title_laws = url + link
r_title_laws = requests.get(url_title_laws)
soup_title_laws = bs(r_title_laws.content)

title_codes_listing =soup_title_laws.find(class_ = 'codes-listing')
title_laws = {}
for a_tag in title_codes_listing.find_all('a'):
    text = a_tag.get_text()
    chapter = text.split('-')[1].strip()
    descript = text.split('-')[2].strip()
    link = a_tag.get('href')
    title_laws[chapter] = [descript,link]

    
df_chapters = pd.DataFrame(title_laws)
df_chapter_laws = df_chapters.T.reset_index()
df_chapter_laws.columns = ['Chapter','Chapter_Description', 'Chapter_Link']
df_chapter_laws.insert(loc = 0,
          column = 'Title',
          value = title)

chapter = '66'
chapter_link = df_chapter_laws['Chapter_Link'][df_chapter_laws['Chapter']== chapter].values[0]
url_chapter_laws = url + chapter_link
r_chapter_laws = requests.get(url_chapter_laws)
soup_chapter_laws = bs(r_chapter_laws.content)

chapter_codes_listing = soup_chapter_laws.find(class_ = 'codes-listing')
chapter_laws = {}
for a_tag in chapter_codes_listing.find_all('a'):
    text = a_tag.get_text()
    section = text.split('-')[2].strip()
    descript = text.split('-')[3].strip()
    link = a_tag.get('href')
    chapter_laws[section] = [descript,link]

df_section = pd.DataFrame(chapter_laws)
df_section_laws = df_section.T.reset_index()
df_section_laws.columns = ['Section','Section_Description', 'Section_Link']
df_section_laws.insert(loc = 0,
          column = 'Chapter',
          value = chapter)
df_section_laws.insert(loc = 0,
          column = 'Title',
          value = title)

section = '1'
url_section_laws = url + df_section_laws['Section_Link'][df_section_laws['Section'] == section].values[0]
r_section_laws = requests.get(url_section_laws)
soup_section_laws = bs(r_section_laws.content)
```
