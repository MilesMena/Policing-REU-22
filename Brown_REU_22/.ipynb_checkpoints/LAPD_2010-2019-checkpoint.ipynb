## Arrests by Week broken down by race

```python
df_2019['Charge Group Description'].unique()
```

```python
def counts_by_date(grouper):
    
    races = df_2019['Descent Code'].unique().tolist()
    date_races_frames = []
    for race in races:
        dataframe = df_2019[df_2019['Descent Code'] == race]
        dataframe = dataframe[dataframe['Charge Group Description'] == grouper]
        count_date_race = dataframe.groupby( by = [pd.Grouper(key = 'DateTime', freq = 'W')])['Counts'].sum().reset_index()
        #count_race = dataframe.groupby(by = grouper)['Counts'].sum().reset_index()
        count_date_race.rename(columns = {'Counts':'counts_%s'%race}, inplace = True)
        #count_race = df_law['Charge Group Description'].reset_index().merge(count_race[['Description','counts_%s'%race]], 
        #                                                      how = 'left', on = 'Description')
        #count_race.drop('index', axis=1, inplace=True)
        date_races_frames.append(count_date_race)
    df_date_counts_race = reduce(lambda left,right: pd.merge(left,right, on = ['DateTime'], how = 'outer'), date_races_frames).fillna(0)
    return df_date_counts_race.replace('Nat','0')
```

```python
charges = df_2019['Charge Group Description'].unique().tolist()
fig, axes = plt.subplots(nrows = len(charges),ncols = 1, figsize = (7,80))
for i,charge in enumerate(charges):
    df_date_counts_race = counts_by_date(charge)
    cols = df_date_counts_race.columns.tolist()[1:]
    if 'DateTime' in cols:
        cols.remove('DateTime')
    weekly_avg = []
    std_weeks = []
    for col in cols:
        weekly_avg.append(df_date_counts_race[col].mean())
        std_weeks.append(df_date_counts_race[col].std())
    #print(weekly_avg)
    #print(std_weeks)
    axes[i].barh(y = cols, width = weekly_avg, xerr = std_weeks, color = 'tomato')
    text = axes[i].set(title = 'Weekly Average Arrests by Race for %s'%charge, xlabel = 'Average Arrests by Week')

fig.tight_layout()
plt.plot()
plt.show()
```

```python
grouper = 'Other Assaults'
fig, ax = plt.subplots(figsize = (10,5))
df_date_counts_race = counts_by_date(grouper)
cols = df_date_counts_race.columns.tolist()[1:]
    #print(df_date_counts_race[cols[2]].unique())
    #print(cols)
weekly_avg = []
std_weeks = []
# calculate mean and std for each race
for col in cols:
    #print(df_date_counts_race[col].mean())
    #print(df_date_counts_race[col].std())
    weekly_avg.append(df_date_counts_race[col].mean())
    std_weeks.append(df_date_counts_race[col].std())
# add our plot to the ith plot
#print(type(weekly_avg),type(std_weeks))
#print(weekly_avg,std_weeks)
ax.barh(y = cols, width = weekly_avg, xerr = std_weeks, color = 'tomato')
text = ax.set(title = 'Weekly Average Arrests by Race for %s'%grouper, xlabel = 'Average Arrests by Week')
```
