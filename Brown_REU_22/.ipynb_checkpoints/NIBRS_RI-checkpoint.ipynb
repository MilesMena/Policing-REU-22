## Import Packages

```python
import pandas as pd
from scipy.fft import rfft, rfftfreq
import numpy as np
from datetime import datetime 
import matplotlib.pyplot as plt
import statistics as stat
from scipy.stats import shapiro
import re
import plotly.graph_objects as go
from statsmodels.tsa.stattools import adfuller
from functools import reduce
plt.style.available

plt.style.use('ggplot')
```

```python
def fast_fourier_transform(counts, what_for):
    normalized_counts = counts/counts.max()
    mean_count = stat.mean(normalized_counts)
    samples = len(normalized_counts)
    sample_rate = 1 # NUMBER OF SAMPLES PER UNIT OF TIME. Might need to change for some samples.
    yf = rfft(normalized_counts - mean_count)
    xf = rfftfreq(samples, 1/sample_rate)  
    max_index = [np.where(np.abs(yf) == np.sort(np.abs(yf))[-(i+1)])[0][0] for i in range(len(np.abs(yf)))]
    #print(max_index)
    for i, index in enumerate(max_index[:5]):
        if index > .001:
            print('Top Frequency %d is %f days' % (i, 1/xf[index]))
    plt.plot(xf, np.abs(yf))
    plt.title('Fast Fourier Transform on %s'% what_for)
    plt.show()
```

```python
df_arrestee_OT = df_arrestee.merge(df_offense_type[['OFFENSE_TYPE_ID','OFFENSE_CODE','OFFENSE_NAME','OFFENSE_CATEGORY_NAME']], how = 'left', on = 'OFFENSE_TYPE_ID')
```

```python
df_single_arrests = df_arrestee_OT[df_arrestee_OT['MULTIPLE_INDICATOR'].isin(['N','C'])]
df_single_arrests['Arrest_count'] = 1
```

```python
#df_single_arrests.groupby(by = ['OFFENSE_CATEGORY_NAME','RACE_ID'])['Arrest_count'].sum().reset_index()
df_single_arrests.groupby(by = 'OFFENSE_CATEGORY_NAME')['Arrest_count'].sum().reset_index()
```

```python
## Try taking each of these data frames and merging it onto the OT_ID
## column of df_offense_type so that we can have a standard length of OT

races = df_single_arrests['RACE_ID'].unique().tolist()
races_frames = []
grouper = 'OFFENSE_CATEGORY_NAME'
for race in races:
    dataframe = df_single_arrests[df_single_arrests['RACE_ID'] == race]
    count_race = dataframe.groupby(grouper)['Arrest_count'].sum().reset_index()
    print(len(count_race))
    count_race.rename(columns = {'Arrest_count':'Arrest_count_%s'%race}, inplace = True)
    print(len(count_race))
    count_race = df_offense_type['OFFENSE_CATEGORY_NAME'].drop_duplicates().reset_index().merge(count_race,
                                                                              how = 'left', on = 'OFFENSE_CATEGORY_NAME')
    print(len(count_race))
    count_race.drop('index', axis=1, inplace=True)
    races_frames.append(count_race)
    
df_counts_race = reduce(lambda  left,right: pd.merge(left,right,on=['OFFENSE_CATEGORY_NAME'],
                                              how='outer'), races_frames).fillna(0)
columns = df_counts_race.iloc[:,1:].columns.tolist()
df_counts_race.dropna(axis = 0, how = 'all', subset = columns, inplace = True)
    
```
